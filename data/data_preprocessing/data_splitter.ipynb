```python
import os
from shutil import copy

project_path = "/home/gatemrou/uds/Eye_Tracking/"
scanpaths = "data/cropped/scanpaths/"
heatmaps = "data/cropped/heatmaps/"
savepath_s = lambda abc, cp: f"data/cropped/scanpaths_sorted/{abc}/{cp}/"
savepath_h = lambda cp: f"data/cropped/heatmaps_sorted/{cp}/"
```

```python
# scanpaths

for filename in os.listdir(project_path+scanpaths):
    im_type = filename[-5]
    im_class = filename[-7]
    copy(project_path+scanpaths+filename, project_path+savepath_s(im_type, im_class)+filename)
```

```python
# heatmaps

for filename in os.listdir(project_path+heatmaps):
    im_class = filename[-7]
    copy(project_path+heatmaps+filename, project_path+savepath_h(im_class)+filename)
```
